```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import mean_squared_error
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import LabelBinarizer
from sklearn.model_selection import cross_val_score
from sklearn.ensemble import RandomForestRegressor
from sklearn.linear_model import ElasticNet
```

```python
org = pd.read_csv('02_trainset.csv')
```

```python
encoder = LabelBinarizer()
c1 = org['C1']
c2 = org['C2']
c3 = org['C3']
c1_1hot = encoder.fit_transform(c1)
c2_1hot = encoder.fit_transform(c2)
c3_1hot = encoder.fit_transform(c3)
c3_1hot.shape
```

```python
c1_1hot_df = pd.DataFrame(c1_1hot, columns=['C1_0', 'C1_1', 'C1_2', 'C1_3', 'C1_4', 'C1_5', 'C1_6', 'C1_7', 'C1_8'])
c2_1hot_df = pd.DataFrame(c2_1hot, columns=['C2_0', 'C2_1', 'C2_2', 'C2_3', 'C2_4', 'C2_5', 'C2_6', 'C2_7'])
c3_1hot_df = pd.DataFrame(c3_1hot, columns=['C3_0', 'C3_1', 'C3_2', 'C3_3', 'C3_4', 'C3_5', 'C3_6', 'C3_7', 'C3_8', 'C3_9', 'C3_10', 'C3_11', 'C3_12', 'C3_13', 'C3_14', 'C3_15', 'C3_16', 'C3_17', 'C3_18', 'C3_19', 'C3_20', 'C3_21', 'C3_22', 'C3_23', 'C3_24', 'C3_25', 'C3_26', 'C3_27', 'C3_28', 'C3_29', 'C3_30', 'C3_31', 'C3_32', 'C3_33', 'C3_34', 'C3_35', 'C3_36', 'C3_37', 'C3_38', 'C3_39', 'C3_40', 'C3_41', 'C3_42', 'C3_43', 'C3_44', 'C3_45', 'C3_46', 'C3_47'])
```

```python
org_1hot = org
org_1hot = org_1hot.drop(['C1', 'C2', 'C3'], axis=1)
org_1hot = org_1hot.join([c1_1hot_df, c2_1hot_df, c3_1hot_df])
```

```python
org_1hot_train = org_1hot
org_1hot_train = org_1hot_train[org_1hot_train.Train_validation_1 == 0]
org_1hot_train
```

```python
train_x = org_1hot_train.drop(['ID', 'Train_validation_1', 'Target1', 'Target2', 'Target3'], axis=1)
train_y = org_1hot_train[['Target1', 'Target2', 'Target3']]
```

```python
forest_reg = RandomForestRegressor(n_estimators=300, n_jobs=-1)
scores = cross_val_score(forest_reg, train_x, train_y, scoring='neg_mean_squared_error', cv= 10)
```

```python
scores.sum()
```

```python
elastic_net = ElasticNet(alpha=0.1, l1_ratio=0.5)
scores = cross_val_score(elastic_net, train_x, train_y, scoring='neg_mean_squared_error', cv= 20)
```
